```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'Company': ['Allianz', 'Diamler', 'Deutsche Bank', 'Continental', 'Brenntag', 'Biersdorf', 'DPDHL', 'Siemens'], 
'Avg. Goals': [55.75, 16.33, 12, 30.5, 35.57, 59, 95.33, 62.5]})
```

```python
fig, ax = plt.subplots(figsize=(12,5))

plt.bar(df['Company'], df['Avg. Goals'])
```

```python
fig, ax = plt.subplots(figsize=(12,6))

plt.bar(df['Company'], df['Avg. Goals'], data=df)
```
