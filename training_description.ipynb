```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_4500 = pd.read_csv('dataset/training_experiment_4500.csv', sep=';')
df_1800 = pd.read_csv('balanced dataset/training_experiment_1800.csv', sep=';')
```

```python
fig, ax = plt.subplots(figsize=(18,6))

ax.plot(df_4500['Step'], df_4500['Training Loss'], label='Training Loss')
ax.plot(df_4500['Step'], df_4500['Validation Loss'], label='Validation Loss')


ax.legend();
print(f'Training with unbalanced dataset')
```
